```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()

llm = ChatGoogleGenerativeAI(model = "gemma-4-31b-it")
```

```python
with open("blood_work.txt","r") as f:
    blood_work = f.read()
```

```python
print(blood_work[:200])
```

```python
extraction_prompt = f"""
    You are a medical data extraction assistant.

    From the blood work report shared below, extract each value and Classify each one as Low, Normal and High 
    based on the reference ranges provided in the report

    Response format:
    Test name: Valur | Status: High / Low / Normal | Reference: Range 

    Blood Report
    {blood_work}
    """

extraction_response = llm.invoke(extraction_prompt)
extraction_text = extraction_response.text

print("STAGE 1: Extracted value")
print(extraction_text)
```

```python
diet_prompt = f"""
    you are a clinical nutritionist who is expert in indian deitery habbits.

    Based on the blood work report provided below, write:
    1. A short summary in 2-3 lines explaining the pateints condition is simple language.
    2. A short practical Indian dietery plan with 2 sections: a) Foods to avoid b) Foods to consume more

    Do not include any other sections in the diet plan..

    Blood Work Report:
    {extraction_text}
"""

diet_response = llm.invoke(diet_prompt)

print("STAGE 2: Health Summary and Diet Plan")
print(diet_response)
```

```python
diet_text = diet_response.text
print(diet_text)
```
