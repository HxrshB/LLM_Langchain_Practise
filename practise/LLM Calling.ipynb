```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model = "gemma-4-31b-it")

response = llm.invoke("How many moons does jupiter have?")

print(response.text)
```

```python
print(response.content)
```

```python
response = llm.invoke(
    [
        ["system","You are a assistant, and answer the query by providing the history of the query."],
        ["human","How many moons does jupiter have?"]
    ]
)
```

```python
print(response.text)
```
